```python
from dv import AedatFile
import numpy as np
```

```python
aedat_file = "../dataset/KH/rain/raw_aedat/2020_07_29_19_28_10.aedat4"
ts = 1596018489623096
tf = 1596018489623096
found = False
with AedatFile(aedat_file) as f:
    for p in f['events'].numpy():
        t = p['timestamp'][0]
        print(t)
        if ts+30*1000000 < t and t < ts+390*1000000:
            if found == False:
                found = True
                e = p
            else:
                e = np.hstack([e,p])
            #print(e.shape)
        elif t >= ts+390*1000000:
            break
print(e.shape)
timestamps, x, y, polarities = e['timestamp'], e['x'], e['y'], e['polarity']
```

```python
ts = timestamps[0]
N = x.shape[0]
print(N)
E = list()
cnt = 0
for i in range(N):
    t = timestamps[i]
    if t - ts >= 1000000:
        ts = t
        Np = np.array(E)
        np.save("../dataset/KH/rain/raw_numpy_1000ms/2020_07_29_19_28_10/"+str(cnt)+".npy",Np)
        Np = None
        E = list()
        cnt += 1
    item = [int(x[i]//2), int(y[i]//2), 1, int((t-ts)//1000)]
    E.append(item)
```
