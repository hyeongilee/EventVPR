```python
import numpy as np
from dv import AedatFile
```

```python
date = "2020_07_31_20_23_29"
env = "night"
ms = 250
utm_file = open("../dataset/KH/"+env+"/utm.txt",'r')
event_file = "../dataset/KH/"+env+"/raw_aedat/"+date+".aedat4"

utm = list()
while True:
    line = utm_file.readline()
    if not line:
        break
    sp_line = line.split()
    utm.append([int(sp_line[0]),float(sp_line[1]),float(sp_line[2])])
utm_file.close()
```

```python
def find_location(timestamp):
    i = 0
    while timestamp > utm[i][0]:
        i += 1
    return utm[i][1], utm[i][2]
```

```python
with AedatFile(event_file) as f:
    events = np.hstack([packet for packet in f['events'].numpy()])
    timestamps, x, y, polarities = events['timestamp'], events['x'], events['y'], events['polarity']
    
ts = timestamps[0]
print(ts)

idx = 0
location_file = open("../dataset/KH/"+env+"/raw_numpy_"+str(ms)+"ms/"+date+"_location_temp.txt",'w')
for t in timestamps:
    if t - ts >= 1000*ms:
        ts = t
        east, north = find_location(t)
        item = str(idx)+' '+str(east)+' '+str(north)+"\n"
        location_file.write(item)
        idx += 1
location_file.close()     
```

```python
location_file = open("../dataset/KH/"+env+"/raw_numpy_"+str(ms)+"ms/"+date+"_location_temp.txt",'r')
location_file_mod = open("../dataset/KH/"+env+"/raw_numpy_"+str(ms)+"ms/"+date+"_location.txt",'w')
cnt = 0
temp = []
prev = [0.0, 0.0]
while True:
    line = location_file.readline()
    if not line:
        break
    sp_line  = line.split()
    cur = [float(sp_line[1]), float(sp_line[2])]
    print(cur)
    item = [int(sp_line[0]),float(sp_line[1]),float(sp_line[2])]
    if prev != cur and cnt != 0:
        dx = cur[0] - prev[0]
        dy = cur[1] - prev[1]
        ddx = dx/cnt
        ddy = dy/cnt
        #print(len(temp),cnt)
        for i in range(cnt):
            temp[i][1] += ddx*i
            temp[i][2] += ddy*i
        for i in range(cnt):
            location_file_mod.write(str(temp[i][0])+' '+str(temp[i][1])+' '+str(temp[i][2])+"\n")
        temp = []
        temp.append(item)
        cnt = 1
        
    else:
        temp.append(item)
        cnt += 1
    prev = cur
location_file.close()
location_file_mod.close()
```
