```python
from dv import AedatFile
import numpy as np
```

```python
env = "night"
date = "2020_07_31_20_23_29"
aedat_file = "../dataset/KH/"+env+"/raw_aedat/"+date+".aedat4"
with AedatFile(aedat_file) as f:
    events = np.hstack([packet for packet in f['events'].numpy()])
    timestamps, x, y, polarities = events['timestamp'], events['x'], events['y'], events['polarity']
```

```python
print(timestamps[0])
```

```python
ts = timestamps[0]
N = x.shape[0]
print(N)
E = list()
cnt = 0
resize_n = 1
ms = 1000

for i in range(N):
    t = timestamps[i]
    if t - ts >= 1000*ms:
        ts = t
        Np = np.array(E)
        print(Np.shape)
        np.save("../dataset/KH/"+env+"/raw_numpy_"+str(ms)+"ms/"+date+'/'+str(cnt)+".npy",Np)
        Np = None
        E = list()
        cnt += 1
    item = [int(x[i]//resize_n), int(y[i]//resize_n), polarities[i], int((t-ts)//1000)]
    E.append(item)
```
