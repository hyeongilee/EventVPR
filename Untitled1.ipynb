```python
import numpy as np
import os
```

```python
def convert_filename(filename_str):
    n = len(filename_str)
    _str = ""
    for i in range(n):
        if filename_str[i] == '_':
            _str += '0'
        else:
            _str += filename_str[i]
    return _str
```

```python
file_location = "/media/student1/770a08ae-0a7d-4b62-bbf4-b81fdd9baf2a/train_dataset/"

txt_file = open(file_location+"test.txt",'r')
while True:
    line = txt_file.readline()
    if not line:
        break
    filename = line.split()[0]
    np_file = np.load(file_location+filename+".npy")
    save_filename = convert_filename(filename)
    print(save_filename)
    np_save = np.save(file_location+save_filename+".npy",np_file)
    os.remove(file_location+filename+".npy")
txt_file.close()
```
