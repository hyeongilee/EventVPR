```python
import numpy as np
import random
from scipy.spatial import distance
```

```python
def merge(database_seq, query_seq):
    N_d = database_seq.shape[0]
    N_q = query_seq.shape[0]
    data = database_seq.tolist()
    temp = query_seq.tolist()
    for i in range(N_q):
        temp[i][2] = 0
    data.extend(temp)
    data = sorted(data, key = lambda x : x[3])
    return data
    
```
